# Task 1:
1.	Create a dataset of hospitalization information for all adults (Age > 18) who received invasive mechanical ventilation. Columns should include HADM_ID, SUBJECT_ID, ADMITTIME, GENDER, Age. Age should be determined based on DOB and ADMITTIME. Assume that documentation of ITEMID = 226732 in the CHARTEVENTS table with value of "Endotracheal tube" indicates that a patient received invasive mechanical ventilation during the hospitalization.

Since Patients has a pesky timestamp rather than a simple 'age', we need to calculate age using TIMESTAMPDIFF(FORMAT,BEING DATE,END DATE), We will need to select this in our select statement so it shows up as a column, but can then use AND statement TIMESTAMPDIFF(YEAR, p.DOB, ADMITTIME) > '18' to sort.

I am joining on SUBJECT_ID in the second join since SUBJECT_ID since HADM_ID is unique to a hospitalization and may leave out measurements.

````python
q1="""SELECT a.HADM_ID, a.SUBJECT_ID, a.ADMITTIME, p.GENDER, TIMESTAMPDIFF(YEAR, p.DOB, ADMITTIME) AS AGE
FROM ADMISSIONS as a JOIN PATIENTS as p ON a.SUBJECT_ID=p.SUBJECT_ID
JOIN CHARTEVENTS as c ON c.SUBJECT_ID=a.SUBJECT_ID
where c.ITEMID='226732'
And TIMESTAMPDIFF(YEAR, p.DOB, ADMITTIME) > '18'
"""
res1=curs.execute(q1)
curs.fetchall()
````

Columns should include HADM_ID, SUBJECT_ID, CHARTTIME, O2_device, tidal_volume. The O2_device column should contain all values of ITEMID = 226732, with missing values at chart times when a tidal volume was documented but not the o2 device. The tidal_volume column should contain all values of ITEMID = 224684, with missing values at chart times when an o2 device was documented but not a tidal volume.

Because this task requires two aliases of the VALUE column in CHARTEVENTS, I am going to make two temporary tables, and Join them on HADM_ID

````python
TT1='''DROP TABLE IF EXISTS t2_; '''
curs.execute(TT1)
O2='''CREATE TEMPORARY TABLE t2_ SELECT c.HADM_ID, c.SUBJECT_ID, c.CHARTTIME, c.VALUE as O2_Device
FROM CHARTEVENTS as c
WHERE c.ITEMID='226732';
'''
curs.execute(O2)
````

````python
TT2='''DROP TABLE IF EXISTS t3_; '''
curs.execute(TT2)
TIDAL='''CREATE TEMPORARY TABLE t3_ SELECT c.HADM_ID, c.SUBJECT_ID, c.CHARTTIME, c.VALUE as tidal_volume
FROM CHARTEVENTS as c
WHERE c.ITEMID='224684';
'''
curs.execute(TIDAL)
````

````python
Task2='''SELECT c.HADM_ID, c.SUBJECT_ID, c.CHARTTIME, t2.O2_Device, t3.tidal_volume
FROM CHARTEVENTS as c JOIN t2_ as t2 ON t2.HADM_ID=c.HADM_ID
JOIN t3_ as t3 on t3.HADM_ID=c.HADM_ID
'''
curs.execute(Task2)
curs.fet()
````
